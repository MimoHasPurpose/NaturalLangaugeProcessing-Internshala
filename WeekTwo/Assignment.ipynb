```python
moontext=open('moon.txt','r',encoding='UTF-8')
# print(moontext.read())
moontexta=moontext.read()
print(type(moontexta))
```

```python
#Task:
"""
ask 1: Annotate the words with part-of-speech tags.

Task 2: Count the number of nouns and verbs in the text.

"""
```

```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(moontexta)
```

```python
doc
```

```python
for token in doc:
  print(token.text,'->',token.pos_,'->',token.tag_)
```

```python
# no of nouns
a = 0
[token.text for token in doc if token.pos_ == 'NOUN']
```
